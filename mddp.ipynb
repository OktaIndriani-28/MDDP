```python
   import numpy as np
   import pandas as pd
   import matplotlib.pyplot as plt
   import seaborn as sns
```

```python
data = pd.read_data = pd.read_excel("/content/sample_data/sample/stroke_data.xlsx")
data.head(300)
```

```python
rentang = [0, 45, 55, np.inf]
kategori = ['Muda', 'Parabaya', 'Tua']
data['age_KATEGORI'] = pd.cut(data['age'], bins=rentang, labels=kategori, include_lowest=True)
data.head(300)
```

```python
rentang = [0, 70, 105, 126, np.inf]
kategori = ['Hipoglikemia', 'Normal', 'Prediabetes ', 'Diabetes']
data['glucose_KATEGORI'] = pd.cut(data['avg_glucose_level'], bins=rentang, labels=kategori, include_lowest=True)
data.head(300)
```

```python
rentang = [0, 1, np.inf]
kategori = ['No', 'Yes']
data['hypertension_KATEGORI'] = pd.cut(data['hypertension'], bins=rentang, labels=kategori, include_lowest=True)
data.head(300)
```

```python
rentang = [0, 18.5, 25, 29.9, 100, np.inf]
kategori = ['under weigth', 'Normal', 'Over weight ', 'Obesitas', 'extreme ']
data['bmi_KATEGORI'] = pd.cut(data['bmi'], bins=rentang, labels=kategori, include_lowest=True)
data.head(300)
```
